### 对时区（tz）进行计数

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec] # 确保rec中有tz
```

#### 取得[‘tz值’:数量]的字典

```python
def get_counts(sequence): # 基础方法
    counts = {} # 创建空字典
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
counts = get_counts(time_zones)
counts['America/New_York']
```

```python
from collections import defaultdict # 高级方法
def get_counts_2(sequence):
    counts = defaultdict(int) # 创建字典，且所有的值均会被初始化为0
    for x in sequence:
        counts[x] += 1
    return counts
counts = get_counts_2(time_zones)
counts['America/New_York']
```

### 利用Python得到前10位的时区

```python
def top_counts(count_dict,n=10):
    value_key_pairs = [(count,tz) for tz,count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
top_counts(counts)
```

```python
from collections import Counter # python标准库中的collections.Counter类可以使上述任务更简单
counts = Counter(time_zones)
counts.most_common(10)
```

#### 按照windows来区分时区

```python
cframe = frame[frame.a.notnull()]
operating_system = np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
operating_system[:5]
```

```python
by_tz_os = cframe.groupby(['tz',operating_system]) # 进行分组
agg_counts = by_tz_os.size().unstack().fillna(0) #size对分组进行计数，用unstack进行重塑
agg_counts[:10]
```

#### 选取最常出现的数组

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

#### 截取最后10行

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh',stacked=True)
```

```python
normed_subset = count_subset.div(count_subset.sum(1),axis=0) # 规范化为“总计为1”的并重新绘图
normed_subset.plot(kind='barh',stacked=True)v
```
